```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)
```

```python
# Dimensions
input_dim = 1
hidden_dim = 1


# Forget gate: f_t = sigmoid(Wf * x_t + Uf * h_{t-1} + bf)
Wf = np.array([[0.5]], dtype=float)
Uf = np.array([[0.1]], dtype=float)
bf = np.array([[0.0]], dtype=float)

# Input gate: i_t = sigmoid(Wi * x_t + Ui * h_{t-1} + bi)
Wi = np.array([[0.4]], dtype=float)
Ui = np.array([[0.2]], dtype=float)
bi = np.array([[0.0]], dtype=float)

# Candidate cell: c~_t = tanh(Wc * x_t + Uc * h_{t-1} + bc)
Wc = np.array([[0.3]], dtype=float)
Uc = np.array([[0.1]], dtype=float)
bc = np.array([[0.0]], dtype=float)

# Output gate: o_t = sigmoid(Wo * x_t + Uo * h_{t-1} + bo)
Wo = np.array([[0.6]], dtype=float)
Uo = np.array([[0.2]], dtype=float)
bo = np.array([[0.0]], dtype=float)

# Initial hidden and cell states
h_t = np.array([[0.0]], dtype=float)   # h_0
c_t = np.array([[0.0]], dtype=float)   # c_0

# ------------------------------------------------------------------
# Final linear layer: y_hat = Wy * h_T + by
# ------------------------------------------------------------------
Wy = np.array([[4.0]], dtype=float)
by = np.array([[0.0]], dtype=float)
```

```python
# Input sequence x = [1, 2, 3, 4]
x_sequence = [1.0, 2.0, 3.0, 4.0]

# Convert each element to a (1,1) column vector
X = [np.array([[x]], dtype=float) for x in x_sequence]

print("Input sequence:", x_sequence)
```

```python
hs = []  # hidden states over time
cs = []  # cell states over time

for t, x_t in enumerate(X, start=1):
    # Forget gate
    f_t = sigmoid(Wf @ x_t + Uf @ h_t + bf)
    # Input gate
    i_t = sigmoid(Wi @ x_t + Ui @ h_t + bi)
    # Candidate cell
    c_t_tilde = tanh(Wc @ x_t + Uc @ h_t + bc)
    # Cell state update
    c_t = f_t * c_t + i_t * c_t_tilde
    # Output gate
    o_t = sigmoid(Wo @ x_t + Uo @ h_t + bo)
    # Hidden state update
    h_t = o_t * tanh(c_t)

    hs.append(h_t.copy())
    cs.append(c_t.copy())

    print(f"Time step {t}")
    print(f"  x_t       = {x_t.ravel()[0]:.4f}")
    print(f"  f_t       = {f_t.ravel()[0]:.4f}")
    print(f"  i_t       = {i_t.ravel()[0]:.4f}")
    print(f"  c~_t      = {c_t_tilde.ravel()[0]:.4f}")
    print(f"  c_t       = {c_t.ravel()[0]:.4f}")
    print(f"  o_t       = {o_t.ravel()[0]:.4f}")
    print(f"  h_t       = {h_t.ravel()[0]:.4f}")
    print("-" * 40)
```

```python
# Last hidden state h_T
h_T = hs[-1]  # shape (1,1)

# Linear mapping to next value
y_hat = Wy @ h_T + by

print(f"Final hidden state h_T: {h_T.ravel()[0]:.4f}")
print(f"Predicted next value y_hat: {y_hat.ravel()[0]:.4f}")
```
