```python
import numpy as np
```

```python
# Step 1: Define token embeddings (7 tokens, d_model = 8)
np.random.seed(42)
tokens = ["What", "are", "the", "symptoms", "of", "diabetes", "?"]
d_model = 8   # embedding dimension
seq_len = len(tokens)
```

```python
# Simulate input embeddings (in practice these come from nn.Embedding)
X = np.random.randn(seq_len, d_model)   # shape: (7, 8)
```

```python
# Step 2: Define weight matrices for Q, K, V
d_k = 8   # key/query dimension
W_Q = np.random.randn(d_model, d_k)    # (8, 8)
W_K = np.random.randn(d_model, d_k)    # (8, 8)
W_V = np.random.randn(d_model, d_k)    # (8, 8)
```

```python
# Step 3: Compute Q, K, V projections
Q = X @ W_Q    # (7, 8)
K = X @ W_K    # (7, 8)
V = X @ W_V    # (7, 8)
```

```python
# Step 4: Compute scaled dot-product attention scores
#   Attention(Q, K, V) = softmax(Q * K^T / sqrt(d_k)) * V
scores = Q @ K.T / np.sqrt(d_k)   # (7, 7)
```

```python
# Step 5: Apply Softmax row-wise to get attention weights
def softmax(x):
    e_x = np.exp(x - np.max(x, axis=-1, keepdims=True))  # numerical stability
    return e_x / e_x.sum(axis=-1, keepdims=True)

attention_weights = softmax(scores)   # (7, 7)
```

```python
# Step 6: Compute the final attended output
output = attention_weights @ V        # (7, 8)

# ─── Print Results ───
print("Tokens:", tokens)
print()
print("Attention Weights (each row = how token attends to others):")
print(np.round(attention_weights, 3))
print()
print("Self-Attention Output (new token representations):")
print(np.round(output, 3))
```
